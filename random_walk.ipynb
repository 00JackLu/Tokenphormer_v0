```python
# Load and pre-process data
from data import get_dataset


adj, features, labels, idx_train, idx_val, idx_test = get_dataset('cora', 3)

# change the format into networkx
G = nx.Graph()
edge_index = adj._indices().numpy()
G.add_edges_from(edge_index.T)   
```

```python
t_num = 5
w_len = 5
```

```python
G = dgl.DGLGraph(G)
```

```python
# nodes_features = torch.empty(G.nodes().size(dim=0), t_num*w_len)
nodes_features = [] 
# 
for i in range(G.nodes().size(dim=0)):
    strat_nodes = [i] * t_num

    token_path = dgl.sampling.random_walk(G, strat_nodes, length=w_len)
    print(token_path[0])
    nodes_features.append(token_path[0])
```
